```python
import math
import cmath
import matplotlib.pyplot as plt
import numpy as np

def sins(x):
    """       ,   )"""
    x_pow = x
    multiplier = 1
    partial_sum = x
    for n in range(1, 20):
        x_pow *= x**2  #     
        multiplier *= -1 / (2*n) / (2*n + 1)  # (-1)^n  
        partial_sum += x_pow * multiplier
    
    return partial_sum

def cotgs(x):
    return math.sqrt(1 - sins(x)**2) / sins(x)

print(sins(2))
print(cotgs(2))

def acotg(x):
    pwx = x
    mult = 1
    sum = x
    for i in range(1, 20):
        pwx *= x**2
        mult *= -1 / (2 * i - 1)
        sum += pwx * mult
    
    return sum

print("112 ",acotg(3))
print("221 ",cotgs(acotg(3)))

%matplotlib inline

angles = np.r_[-16.25:16.25:0.01]
plt.plot(angles, np.sin(angles))
plt.plot(angles, np.vectorize(sins)(angles))
plt.show()

%matplotlib inline

angles = np.r_[-5.25:5.25:0.1]
plt.plot(angles, np.cos(angles))
plt.plot(angles, np.vectorize(cotgs)(angles))
plt.show()
#    

%matplotlib inline

angles = np.r_[-3.75:3.75:0.01]
plt.plot(angles, np.vectorize(acotg)(angles))
plt.plot(angles, np.arctan(angles))
plt.show()
#     ,   ,  
```
